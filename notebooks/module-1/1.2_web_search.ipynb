## Without web search

```python
import os
from dotenv import find_dotenv, load_dotenv

dotenv_path = find_dotenv(usecwd=True)
if not dotenv_path:
    raise FileNotFoundError("Could not find a .env file from the current working directory")

load_dotenv(dotenv_path, override=True)
print(f"Loaded environment from: {dotenv_path}")
print(f"TAVILY_API_KEY loaded: {bool(os.getenv('TAVILY_API_KEY'))}")
```

```python
from langchain.agents import create_agent

agent = create_agent(
    model="gpt-5-nano"
)
```

```python
from langchain.messages import HumanMessage

question = HumanMessage(content="How up to date is your training knowledge?")

response = agent.invoke(
    {"messages": [question]}
)
```

```python
print(response['messages'][-1].content)
```

## Add web search tool

```python
from langchain.tools import tool
from typing import Dict, Any
from tavily import TavilyClient

tavily_client = TavilyClient()

@tool
def web_search(query: str) -> Dict[str, Any]:

    """Search the web for information"""

    return tavily_client.search(query)

web_search.invoke("Who is the current CM of kerala?")
```

```python
agent = create_agent(
    model="gpt-5-nano",
    tools=[web_search]
)

question = HumanMessage(content="Who is the current mayor of San Francisco?")

response = agent.invoke(
    {"messages": [question]}
)
```

```python
from pprint import pprint

pprint(response['messages'])
```
